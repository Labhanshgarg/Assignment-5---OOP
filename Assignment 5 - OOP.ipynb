## Write a Python class to implement pow(x, n)



```python
class power:
    def __init__(self,x,n):
        self.x=x
        self.n=n
        self.result=0
    def nth_power(self):
        self.result=pow(self.x,self.n)
        print(self.result)
```

```python
find=power(10,2)
```

```python
find.nth_power()
```

```python
find=power(10,3)
```

```python
find.nth_power()
```
